# Task
Calculate the total number of times each player was targeted by summing the 'pass_attempt' column for each receiver across all plays in the `giants_off` DataFrame where a receiver was identified, regardless of whether the pass was completed or not.

## Calculate Total Targets per Receiver

### Subtask:
Calculate the total number of times each player was targeted by summing the 'targets' column for each receiver across all plays (completed or incomplete) where a receiver was identified.


**Reasoning**:
To calculate the total number of targets per receiver, I will group the 'giants_off' DataFrame by 'receiver_player_name' and sum the 'pass_attempt' column.



```python
total_targets = (
    giants_off
    .groupby("receiver_player_name")["pass_attempt"]
    .sum()
    .reset_index(name="total_targets")
)
total_targets
```

## Calculate Receptions, Receiving Yards, and EPA for Completed Passes

### Subtask:
Filter the `giants_off` DataFrame to include only completed passes. Then, group by `receiver_player_name` to calculate the sum of `complete_pass` (for total receptions), sum of `receiving_yards` (for total receiving yards), and the mean of `epa` (for average EPA per reception).


**Reasoning**:
To calculate the total receptions, receiving yards, and average EPA per reception, I will filter the 'giants_off' DataFrame for completed passes, then group by 'receiver_player_name' and apply the specified aggregations.



```python
completed_passes = giants_off[giants_off['complete_pass'] == 1]
completed_pass_stats = (
    completed_passes
    .groupby("receiver_player_name")
    .agg(
        receptions=("complete_pass", "sum"),
        receiving_yards=("receiving_yards", "sum"),
        avg_epa_per_reception=("epa", "mean")
    )
    .reset_index()
)
completed_pass_stats.head()
```

**Reasoning**:
To get a comprehensive view of receiver performance, including catch rate, I will merge the 'total_targets' DataFrame (which contains total targets per receiver) with the 'completed_pass_stats' DataFrame (which contains receptions, receiving yards, and EPA for completed passes).



```python
receiver_stats = pd.merge(total_targets, completed_pass_stats, on='receiver_player_name', how='left')
receiver_stats['catch_rate'] = receiver_stats['receptions'] / receiver_stats['total_targets']
receiver_stats = receiver_stats.fillna(0)
receiver_stats.sort_values(by='total_targets', ascending=False).head()
```

**Reasoning**:
To visualize the catch rates for the top 10 receivers, I will create a bar chart using the `receiver_stats` DataFrame, sorting it by `total_targets` and then selecting the top 10.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort by total_targets and get the top 10 receivers
top_10_receivers = receiver_stats.sort_values(by='total_targets', ascending=False).head(10)

# Create the bar chart
plt.figure(figsize=(12, 7))
sns.barplot(x='receiver_player_name', y='catch_rate', data=top_10_receivers, palette='viridis')
plt.title('Top 10 Receivers by Catch Rate (Sorted by Total Targets)')
plt.xlabel('Receiver Player Name')
plt.ylabel('Catch Rate')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` because the `palette` argument was used without assigning a `hue` variable. To fix this and follow best practices, I will assign the `x` variable to `hue` and set `legend=False` in the `sns.barplot` call.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort by total_targets and get the top 10 receivers
top_10_receivers = receiver_stats.sort_values(by='total_targets', ascending=False).head(10)

# Create the bar chart
plt.figure(figsize=(12, 7))
sns.barplot(x='receiver_player_name', y='catch_rate', data=top_10_receivers, hue='receiver_player_name', palette='viridis', legend=False)
plt.title('Top 10 Receivers by Catch Rate (Sorted by Total Targets)')
plt.xlabel('Receiver Player Name')
plt.ylabel('Catch Rate')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
