```python
giants_off["season"].value_counts()
giants_off["week"].value_counts().sort_index()
```

```python
giants_points = (
    giants_off
    .groupby("game_id")["posteam_score_post"]
    .max()
    .reset_index(name="points_scored")
)
```

```python
giants_points
```

```python
ppg = giants_points["points_scored"].mean()
ppg
```

Average points score through week 15 is 21.6875 points.

```python
giants_off_epa = giants_off["epa"].mean()
giants_off_epa
```

```python
run_pass_epa = (
    giants_off
    .groupby("play_type")["epa"]
    .mean()
    .reset_index()
)
run_pass_epa
```

```python
run_pass_epa = run_pass_epa[
    run_pass_epa["play_type"].isin(["run", "pass"])
]
```

```python
run_pass_epa
```

```python
qb_epa = (
    giants_off
    .groupby("passer_player_name")
    .agg(
        plays=("play_id", "count"),
        avg_epa=("epa", "mean"),
        total_epa=("epa", "sum"),
        cpoe=("cpoe", "mean")
    )
    .reset_index()
    .sort_values("total_epa", ascending=False)
)
qb_epa
```

```python
receivers = (
    giants_off[giants_off["complete_pass"] == 1]
    .groupby("receiver_player_name")
    .agg(
        targets=("pass_attempt", "count"),
        receptions=("complete_pass", "count"),
        yards=("receiving_yards", "count"),
        avg_epa=("epa", "mean")
    )
    .reset_index()
    .sort_values("yards", ascending=False)
)
receivers
```
